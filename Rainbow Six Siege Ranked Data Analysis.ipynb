```python
grouped_data = df.groupby("mapname")
```

Let's see the number of role wins that each map had.

First off is the attacker role.

```python
attacker_win_data = grouped_data["winrole"].apply(lambda x: (x == "Attacker").sum())
print(attacker_win_data)
```

Now for the defender role.

```python
defender_win_data = grouped_data["winrole"].apply(lambda x: (x == "Defender").sum())
print(defender_win_data)
```

Merging the two dataframes into one.

```python
merged_data = df.groupby("mapname")["winrole"].value_counts().unstack(fill_value=0)
merged_data.columns = ["Attacker", "Defender"]
merged_data
```

Let's now find the ratio between the two roles, and append that to the data frame

```python
merged_data["Ratio"] = merged_data["Attacker"] / merged_data["Defender"]
merged_data.columns = ["Attacker", "Defender", "Ratio"]
```

```python
merged_data
```

Let's categorize the maps and determine if the maps are attacker or defender sided.

```python
def categorize_map(ratio):
    if ratio > 1.05:
        return "Attacker"
    elif ratio < 0.95:
        return "Defender"
    else:
        return "Balanced"
```

If the ratio is greater than 1.05, it is an attacker sided map. <br>
If the ratio is less than 0.95, it is a defender sided map. <br>
Else, the map is reasonably balanced

```python
merged_data["A/D Sided Map"] = merged_data["Ratio"].apply(categorize_map)
merged_data.columns = ["Attacker", "Defender", "Ratio", "A/D Sided Map"]
```
